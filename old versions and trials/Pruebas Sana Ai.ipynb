Modelo 600M Parametros a 1024px

```python
import torch
from diffusers import SanaPipeline
import gradio as gr

pipe = SanaPipeline.from_pretrained(
    "Efficient-Large-Model/Sana_600M_1024px_diffusers",
    variant="fp16",
    torch_dtype=torch.float16,
)
pipe.to("cuda")

pipe.vae.to(torch.bfloat16)
pipe.text_encoder.to(torch.bfloat16)





```

```python
def generate_image(prompt: str):

   # prompt = 'a cyberpunk cat with a neon sign that says "Sana"'
    image = pipe(
        prompt=prompt,
        height=1024,
        width=1024,
        guidance_scale=5.0,
        num_inference_steps=20,
        generator=torch.Generator(device="cuda").manual_seed(42),
    )[0]

    image[0].save("sana.png")
    return 'sana.png'
```

```python
import gradio as gr


gr.Interface(
    fn=generate_image,
    inputs=gr.Textbox(lines=2, placeholder="Escribe tu prompt aquí..."),
    outputs=gr.Image(type="filepath"),  # Se retorna el path del archivo de imagen
    title="Generador de Imágenes con Sana",
    description="Ingresa un prompt y el modelo Sana generará la imagen correspondiente en formato PNG.",
).launch(debug=True)
```
